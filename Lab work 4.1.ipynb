## Завдання 1:
- Підключитися до API Національного банку.
- Отримати курс доллара за останній рік.
- Вивести середнє значення та відхилення курсу за кожен місяць.
- Дану інформацію записати у файл за допомогою pickle

```python
import requests
import datetime
import statistics
import pickle
```

```python
#Підключитися до API Національного банку.
today_date = datetime.date.today()
rates = {}
for i in range(0,366):
    previous_date = today_date - datetime.timedelta(i)
    params = {'date': previous_date.strftime('%Y%m%d')}
    
    r = requests.get('https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange?valcode=USD&json', params).json()
    rates[previous_date] = r[0]['rate']
```

```python
rates
```

```python
#Отримати курс доллара за останній рік.
average_rate = {}
standart_deviation = {}
for m in range(1,13):
    rates_month = [rates[i] for i in rates.keys() if i.month==m]
    average_rate[m] = statistics.mean(rates_month)
    standart_deviation[m] = statistics.stdev(rates_month)

print('Середнє значення курсу за місяцями:')
for el in average_rate.items():
    print(el)
print()
print('Середньоквадратичне відхилення курсу за місяцями:')
for el in standart_deviation.items():
    print(el)

```

```python
#Дану інформацію записати у файл за допомогою pickle
with open('rates.pickle', 'wb') as f:
    pickle.dump(rates, f)
    pickle.dump(average_rate, f)
    pickle.dump(standart_deviation, f)
```

```python
with open('rates.pickle', 'rb') as f:
    rates_f = pickle.load(f)
    average_rate_f = pickle.load(f)
    standart_deviation_f = pickle.load(f)
print(average_rate_f)
print(standart_deviation_f)
```
