Програма приймає на вхід два параметри: стать спортсмена(`sex = (f, m)`) та вік спортсмена(`age = [0,100]`).  Спортсмен-дівчина може брати участь у змаганні, якщо її вік більший 18, спортсмен-хлопець - якщо вік більший 16

Якщо введеніть не коректні дані, виведіть повідомлення про помилку

```python
try:
    sex = input('Cтать спортсмена - ')
    age = float(input('Вік спортсмена - '))
    if 'm' in sex:
        if 100 >= age >=18:
            print('Спортсмен може брати участь у змаганні')
        else:
            print('Спортсмен не може брати участь у змаганні')
    elif 'f' in sex:
        if 100 >= age >= 16:
            print('Спортсмен може брати участь у змаганні')
        else:
            print('Спортсмен не може брати участь у змаганні')
    else:
        print('Спробуйте ще раз')
except ValueError:
    print('Введіть число')
    print('Спробуйте ще раз')
except Exception:
    print('Спробуйте ще раз')
```

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,10) for i in range(1000)]
```

```python
# Рохраховує мінімальне та максимальне значення
print('Мінімальне значення -', min(list_with_random_numbers))
print('Максимальне значення -', max(list_with_random_numbers))

# Середнє значення послідовності
print('Cереднє значення послідовності -', sum(list_with_random_numbers)/len(list_with_random_numbers))

# Медіану послідовності
from statistics import median
print('Медіана послідовності -', median(list_with_random_numbers))

# Частоту кожної цифри в послідовності
numbers_counter = {}
for number in list_with_random_numbers:
    if number not in numbers_counter:
        numbers_counter[number] = 1
    else:
        numbers_counter[number] += 1
print('Частота кожної цифри в послідовності:')
for i,number in enumerate(sorted(numbers_counter,key=numbers_counter.get,reverse=True)):
    print(number, '-', numbers_counter[number])

# Середнє квадратичне відхилення
#import numpy as np
#print(np.std(list_with_random_numbers))
from math import sqrt
print('Середнє квадратичне відхилення -', sqrt(sum((x-sum(list_with_random_numbers)/len(list_with_random_numbers))**2 for x in list_with_random_numbers)/(len(list_with_random_numbers))))
```
